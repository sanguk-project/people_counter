```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches
import seaborn as sns
```

```python
# Excel 파일 불러오기
df1 = pd.read_excel('test1LR_analysis.xlsx')
df2 = pd.read_excel('test2UD_analysis.xlsx')
```

```python
df2
```

```python
# 'device' 컬럼에서 '3090Ti'인 행만 선택
df1_3090ti = df1[df1['device'] == '3090ti']
df1_3090ti
```

```python
# Set style
sns.set(style='whitegrid')

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Data for two plots
plots = [('fps (fps)', axes[0]), ('counting_accuracy (%)', axes[1])]

# Plotting
for plot in plots:
    unit = plot[0].split(' ')[0].lower() # Extract the unit and convert it to lower case
    ax = sns.barplot(x='precision', y=unit, hue='img_size', data=df1_3090ti, ax=plot[1], errorbar=None, order=['fp32', 'fp16', 'int8'])
    ax.set(ylabel=plot[0]) # Set y-label with unit

    # Adding data values on top of each bar
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width() / 2, p.get_height(), '{0:.2f}'.format(p.get_height()), fontsize=12, ha='center', va='bottom')

    # Move the legend to an empty part of the plot and set the title
    ax.legend(loc='upper right', bbox_to_anchor=(1.01, 1.21), title='img_size')

# Set the overall title
plt.suptitle('NVIDIA GeForce RTX 3090Ti (Test1)', fontsize=18)

# Adjust layout for better view
plt.tight_layout()
plt.show()
```

```python
# Set style
sns.set(style='whitegrid')

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15, 15))

# Data for two plots
plots = [('fps (fps)', axes[0]), ('counting_accuracy (%)', axes[1])]

# Define unique categories in 'img_size'
img_sizes = df1_3090ti['img_size'].unique()

# Generate a colormap
cmap = plt.get_cmap("tab10")

# Define custom rectangles for legend
legend_rects = [matplotlib.patches.Patch(color=cmap(i), label=label) for i, label in enumerate(img_sizes)]

# New labels for y-axis
y_labels = ["FPS (fps)", "Counting Accuracy (%)"]

# Capitalize precision labels
df1_3090ti.loc[:, 'precision'] = df1_3090ti['precision'].str.upper()

# Plotting
for i, plot in enumerate(plots):
    unit = plot[0].split(' ')[0].lower() # Extract the unit and convert it to lower case
    ax = sns.barplot(x='precision', y=unit, hue='img_size', data=df1_3090ti, ax=plot[1], palette='tab10', errorbar=None, order=['FP32', 'FP16', 'INT8'])
    ax.set(ylabel=y_labels[i], xlabel=None) # Set y-label with unit and remove x-label
    ax.xaxis.label.set_size(20) # Set x-axis label size
    ax.yaxis.label.set_size(20) # Set y-axis label size
    ax.tick_params(axis='x', labelsize=15) # Set x-axis tick label size
    ax.tick_params(axis='y', labelsize=15) # Set y-axis tick label size

    # Adding data values on top of each bar
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width() / 2., p.get_height(), '{0:.2f}'.format(p.get_height()), fontsize=15, ha='center', va='bottom')

    # Remove default seaborn legend
    ax.get_legend().remove()

# Set the overall title
plt.suptitle('NVIDIA GeForce RTX 3090Ti (Test1)', fontsize=25)

# Create an overarching legend on top of the plots
plt.figlegend(handles=legend_rects, loc='upper right', ncol=len(legend_rects), title='Image Size')

# Adjust layout for better view
plt.tight_layout()
plt.subplots_adjust(top=0.9) # Adjust the subplot to fit the figlegend
plt.show()
```

### Input Image Size에 따른 성능비교 자료
